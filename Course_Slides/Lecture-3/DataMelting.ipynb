### 1) Column headers are values, not variable names

```python
#Importing data

df = pd.read_excel("Data_Extract_From_World_Development_Indicators.xlsx")
```

```python
#Exploring data

df.head()
```

```python
#Melting function

d = pd.melt(df, id_vars = ['Country Name','Country Code'], var_name='Date', value_name='GDPperCapGrowth%')
```

```python
d
```

```python
#Reshape the data by giving index/column 
c = d.pivot(index = "Country Name", columns = 'Date' ).drop("Country Code", axis = 1)
c.head()
```
